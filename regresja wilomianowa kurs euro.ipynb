```python
import operator

import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures


x=[[4.02],[4.13],[4.02],[3.91],[3.83],[3.94],[3.95],[3.84],[3.88],[3.80],[3.78],[3.65],[3.57],[3.41],[3.31],[3.77],[4.49],[4.44],
   [4.19],[4.17],[3.98],[4.01],[4.01],[3.96],[3.94],[3.95],[4.15],[4.42],[4.22],[4.26],[4.13],[4.11],[4.15],[4.19],[4.24],[4.18],
   [4.18],[4.16],[4.17],[4.21],[4.19],[4.08],[4.18],[4.26],[4.36],[4.37],[4.34],[4.37],[4.32],[4.21],[4.25],[4.23],[4.18],[4.26],
   [4.30],[4.29]]
y=[[56405],[53648],[55942],[56086],[58426],[57231],[56402],[56984],[73694],[70154],[73158],[75642],[65473],[82351],[80654],
   [81004],[29313],[28720],[70526],[74256],[56488],[54867],[50274],[58463],[67610],[69213],[63622],[74148],[77577],[71236],
   [57697],[66701],[75906],[72085],[75816],[77176],[52522],[77795],[69525],[82517],[92108],[86412],[80826],[96991],[105236],
   [106486],[92850],[113486],[127098],[122628],[109003],[127606],[139885],[133160],[130298],[128546]]
polynomial_features= PolynomialFeatures(degree=2)
x_poly = polynomial_features.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)
y_poly_pred = model.predict(x_poly)

rmse = np.sqrt(mean_squared_error(y,y_poly_pred))
r2 = r2_score(y,y_poly_pred)
print(rmse)
print(r2)

plt.scatter(x, y, s=10)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred), key=sort_axis)
x, y_poly_pred = zip(*sorted_zip)
plt.plot(x, y_poly_pred, color='m')
plt.title('Polynomial Regression') 
plt.xlabel('kurs euro')
plt.ylabel('sprzedaż samochodów')
plt.show()
```
